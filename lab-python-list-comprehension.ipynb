````python
def get_costumer_orders():
    order = int(input("How many products do you want?"))
    costumer_orders = {(input(f"What product do you want?")) for i in range (order)}
    return costumer_orders
````

````python
costumers_orders = get_costumer_orders()
````

````python
print(costumers_orders)
````

````python
def total_price(costumer_orders):
    cost_of_products = [float(input(f"How much is it?")) for i in costumer_orders]
    result = sum(cost_of_products)
    return result
````

````python
total_price(costumer_orders)
````

````python
costumer_orders = get_costumer_orders()
````

````python
def update_inventory(costumers_orders,inventory):
    updated_inventory = {product:(qt - 1 if product in costumer_orders else qt) for product,qt in inventory.items()}
    new_inventory = {product: qt for product, qt in updated_inventory.items() if qt > 0}
    
    return new_inventory
````

````python
update_inventory(costumers_orders,inventory)
````
